### Mutabilidad: una distinción importante

Una idea clave en Python es diferenciar objetos **mutables** e **inmutables**:

- **mutables**: pueden cambiar después de ser creados, como `list` y `dict`;
- **inmutables**: no cambian, como `tuple`, `int`, `float` y `str`.

Esta diferencia ayuda a entender muchos comportamientos del lenguaje.

## Indexación y *slicing*

Las reglas para acceder a elementos individuales o subconjuntos de listas, tuplas o arrays de `NumPy` son:


````python
 #a[ini:fin] # items desde ini hasta fin-1
 # a[ini:]    # items desde ini hasta el resto del array
 # a[:fin]    # items desde el inicio hasta el final -1
 # a[:]       # todo el array`
````

También se puede usar `step` con cualquiera de las expresiones anteriores:


`a[ini:fin:step] #empieza en ini hasta fin-1 separados por step`

La idea clave es recordar que la indexación empieza en **0**, no en 1.  
Además, `:fin` representa el **primer índice que no se incluye**.

Los índices también pueden ser negativos, lo que indica que se cuenta desde el final del objeto.


````python
#a[-1] #ultimo ítem del array
#a[-2:] #dos ultimos ítems del array
#a[:-2] #todo excepto los dos últimos ítems`
````

También `a[:5]` devuelve los 5 primeros elementos, y `a[-5:]` los últimos.

### Condicionales


**`if`**



Ejecuta un bloque de sentencias (que deben estar **indentadas**) si la condición es verdadera.

También se puede usar `else`:



Definamos una función simple para evaluar condicionales:


````python
def signo_de_a(a):
    if(a < 0.0):
        signo = 'negativo'
    elif(a>0.0):
        signo = 'positivo'
    else: 
        signo = 'zero'
    return(signo)
            
````

### Bucles


**`while`**


```python
while condicion:
    bloque
```


Ejecuta el bloque de sentencias mientras la condición sea verdadera.  
Después de cada iteración, la condición se vuelve a evaluar.


````python
nMax = 5
n = 1
a = [] # creamos una lista vacía
while n < nMax:
    a.append(1.0/n) # agregamos un elemento a la lista
    n += 1
print(a)
````

## `for`


Esta sentencia necesita una variable objetivo y una secuencia sobre la que iterar. Su forma general es:


```python
for objetivo in secuencia:
    bloque
```


También es posible añadir un `else` después del `for` si es necesario (y aplica también para `while`).


````python
nMax = 5
a = []
for n in range(1,nMax):
    a.append(1.0/n)
print(a)
````

Aquí `n` es la variable objetivo y la secuencia se crea con la función `range`.  
Veamos un ejemplo más completo que busca un nombre en una lista:


````python
lista = ["Julio","Jammie","Gina","David"]
name = "Tito"#eval(input("Escriba un nombre: ")) #Codigo para ingresar datos
for i in range(len(lista)):
    if lista[i] == name:
        print( name, 'es número', i+1, 'en la lista')
        break
else:
    print( name, 'no está en la lista')
````

### `continue`


La sentencia `continue` permite saltar el resto del código dentro de una iteración y pasar inmediatamente a la siguiente.

El siguiente ejemplo construye una lista con todos los números entre 1 y 99 divisibles para 7.


### Sugerencia de estilo

En muchos casos, cuando solo necesitas iterar sobre una secuencia, es preferible:

- usar `for elemento in secuencia` en lugar de índices;
- usar `enumerate(...)` cuando además necesitas la posición.


````python
x = [] # creamos una lista vacía
for i in range(1,100):
    if i%7!=0: continue # Si no es divisible para 7, no se ejecuta el resto
    x.append(i) # agregamos i a la lista
print(x)
````

## Programas en Python
### Funciones, módulos y paquetes


- **Función**: se define con la palabra clave `def` y normalmente devuelve un resultado con `return`.
- **Módulo**: es un archivo con extensión `.py` que contiene funciones, clases o variables.
- **Paquete**: es una carpeta que agrupa módulos relacionados; normalmente incluye un archivo `__init__.py`.


**Funciones**

````python
'''Demostración de una función de python
author: thomas haslwanter, date: May-2015
'''

# Importamos los paquetes
import numpy as np

def incomeAndExpenses(data):
    '''Encuentre la suma de valores positivos, y la suma de valores negativos.'''
    income = np.sum(data[data>0])
    expenses = np.sum(data[data<0])
    return(income,expenses)

````

````python
testData = np.array([-5,12, 3, -6, -4, 8])
(myIncome,myExpenses) = incomeAndExpenses(testData)
print('Has ganado USD {0:5.2f} y gastado USD {1:5.2f}'.format(myIncome,-myExpenses))
````

El número de parámetros en la definición de una función puede ser arbitrario. Por ejemplo, en la definición:


`def func(x1,x2,*x3)`

`x1` y `x2` son parámetros posicionales.  
En cambio, `*x3` representa una tupla de longitud arbitraria que agrupa argumentos adicionales.

Si llamamos a la función `func(a, b, c, d, e)`, entonces:

- `a` corresponde a `x1`
- `b` corresponde a `x2`
- `c, d, e` quedan agrupados dentro de `x3`


### Funciones `lambda`


Si una función puede expresarse en una sola línea, se puede definir con `lambda`:

```python
func_name = lambda param1, param2, ...: expresion
```

Solo se permite **una expresión**. Un ejemplo:


````python
c = lambda x,y: x**2 + y**2
print(c(3,4))
````
